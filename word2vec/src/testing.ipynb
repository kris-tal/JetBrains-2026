```python
import numpy as np
from utils import load_data, split_data, Vocabulary
from model import CBOW
```

```python
data_path = "../data/emotion_dataset.csv"
df_tokenized = load_data(data_path)
```

```python
df_subset = df_tokenized.sample(frac=0.15, random_state=42)
```

```python
training_set, test_set = split_data(df_subset, training_ratio=0.8)
```

```python
vocab = Vocabulary(training_set, min_count=3)
print(vocab.size)
```

```python
print(vocab.size)
print(vocab.word2idx)
```

```python
model = CBOW(training_set, vocab, window_size=2, embedding_dim=60, epochs=10, learning_rate=0.01, neg_sample_size=7)
model.fit()
```

```python
model.evaluate(test_set)
```
